### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
SL=round(Iris['Sepal Length (in cm)'].max()-Iris['Sepal Length (in cm)'].min(),2)
SW=round(Iris['Sepal Width in (cm)'].max()-Iris['Sepal Width in (cm)'].min(),2)
PL=round(Iris['Petal length (in cm)'].max()-Iris['Petal length (in cm)'].min(),2)
PW=round(Iris['Petal width (in cm)'].max()-Iris['Petal width (in cm)'].min(),2)
print('Range of Sepal Length (in cm):',SL)
print('Range of Sepal Width (in cm):',SW)
print('Range of Petal Length (in cm):',PL)
print('Range of Petal Width (in cm):',PW)
```
